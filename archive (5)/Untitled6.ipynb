```python
from types import new_class
import pandas as pd
from sklearn.linear_model import LinearRegression

df=pd.read_csv("/content/london_houses.csv")
x=df[['Square Meters','Bedrooms','Building Age']]
y=df['Price (£)']

cv=LinearRegression()
cv.fit(x,y)
print("Size weight(a):",cv.coef_[0])
print("Bedroom weight(b):",cv.coef_[1])
print("Age weight(d):",cv.coef_[2])
print("constant(C):",cv.intercept_)

size=int(input("Enter the size of the house:"))
bedroom=int(input("Enter the number of bedrooms:"))
age=int(input("Enter the age of the house:"))

new_data=pd.DataFrame([[size,bedroom,age]],columns=['Square Meters','Bedrooms','Building Age'])
predicted_price=cv.predict(new_data)
print("Predicted house price:",predicted_price)
```
